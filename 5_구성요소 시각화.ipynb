```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#  
data = {
    "Seller Grade": ["Diamond", "Gold", "Silver", "Bronze", "Iron"],
    "Product Description Length": [715.46, 711.31, 688.50, 659.37, 650.79],
    "Product Name Length": [48.20, 47.77, 46.78, 45.82, 44.96],
    "Number of Product Images": [2.19, 2.12, 2.12, 2.06, 2.04]
}

df = pd.DataFrame(data)

#   
sns.set(style="whitegrid")
plt.rc('font', family = 'Malgun Gothic')
plt.figure(figsize=(15, 12))

# 1.   
plt.subplot(3, 1, 1)
ax1 = sns.barplot(x="Seller Grade", y="Product Description Length", data=df, palette="coolwarm")
plt.title("     ")
plt.ylim(640, 720)  #    
for i, val in enumerate(df["Product Description Length"]):
    ax1.text(i, val + 1, f"{val:.1f}", ha='center', va='bottom')
plt.ylabel("  ")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 2.   
plt.subplot(3, 1, 2)
ax2 = sns.barplot(x="Seller Grade", y="Product Name Length", data=df, palette="viridis")
plt.title("     ")
plt.ylim(44, 49)  #   
for i, val in enumerate(df["Product Name Length"]):
    ax2.text(i, val + 0.1, f"{val:.2f}", ha='center', va='bottom')
plt.ylabel("  ")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 3.   
plt.subplot(3, 1, 3)
ax3 = sns.barplot(x="Seller Grade", y="Number of Product Images", data=df, palette="magma")
plt.title("     ")
plt.ylim(0.00, 7.00)
for i, val in enumerate(df["Number of Product Images"]):
    ax3.text(i, val + 0.005, f"{val:.2f}", ha='center', va='bottom')
plt.ylabel("  ")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
